```python
import pandas as pd
import altair as alt
```

# 1. Introducción de Deportes en los Juegos Olímpicos

```python
df = pd.read_csv('tarea1-1.csv')
```

```python
df['rank'] = df.groupby('Debut').cumcount() + 1
df['year'] = df['Debut'].str.extract(r'(\d{4})').astype(int)

debut_order = (df.drop_duplicates('Debut').sort_values('year')['Debut'].tolist())

chart = alt.Chart(df).mark_circle(size=90, opacity=0.85).encode(
    x=alt.X('Debut:N',
            sort=debut_order,
            title='Juegos',
            axis=alt.Axis(labelAngle=40)),
    y=alt.Y('rank:Q',
            title='Deportes introducidos',
            axis=alt.Axis(tickMinStep=1)),
    color=alt.Color('Debut:N', legend=None),
    tooltip=[
        alt.Tooltip('Sport:N', title='Deporte'),
        alt.Tooltip('Debut:N', title='Juegos'),
    ]
).properties(
    title='Introducción de Deportes en los JJ.OO.',
    width=750,
    height=320,
)
```
