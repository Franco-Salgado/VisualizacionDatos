```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Cargar CSV
df = pd.read_csv("../../sources/data-ia/JJ.OO. (Respuestas) - Respuestas de formulario 1.csv")

# Columna de deportes favoritos
deporte_col = "¿Qué deporte de los JJ.OO. disfrutas más ver? "

# Separar respuestas múltiples
todos_deportes = []

for respuesta in df[deporte_col].dropna():
    deportes = [d.strip() for d in str(respuesta).split(",")]
    todos_deportes.extend(deportes)

# Contar frecuencia
conteo = Counter(todos_deportes)

# Convertir a DataFrame
conteo_df = pd.DataFrame(
    conteo.items(),
    columns=["Deporte", "Cantidad"]
).sort_values(by="Cantidad", ascending=False)

# Graficar
plt.figure(figsize=(8,5))
plt.bar(conteo_df["Deporte"], conteo_df["Cantidad"])
plt.title("Deportes más populares entre los encuestados")
plt.xlabel("Deporte")
plt.ylabel("Cantidad de menciones")
plt.xticks(rotation=45)
plt.show()
```
